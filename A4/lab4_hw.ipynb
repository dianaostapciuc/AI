```python
# exercise 1:
import torch
import torch.nn.functional as F
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

# derivative:
def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))

points = [0.0, 1.0]
for point in points:
    x = torch.tensor(point, requires_grad=True, dtype=torch.float)
    y = sigmoid(x)
    y.backward()
    pytorch_gradient = x.grad.item()
    
    explicit_gradient = sigmoid_gradient(x).item()
    
    print(f"Point: {point}")
    print("PyTorch Gradient:", pytorch_gradient)
    print("Explicit Formula Gradient:", explicit_gradient)
    print()
```

```python
# exercise 2:
import torch

def linear(x):
    return x

# derivative
def linear_gradient(x):
    return torch.ones_like(x)

points = [0.0, 1.0]
for point in points:
    x = torch.tensor(point, requires_grad=True, dtype=torch.float)
    y = linear(x)
    y.backward()
    pytorch_gradient = x.grad.item()
    
    explicit_gradient = linear_gradient(x).item()
    
    print(f"Point: {point}")
    print("PyTorch Gradient:", pytorch_gradient)
    print("Explicit Formula Gradient:", explicit_gradient)
    print()
```

```python
# exercise 3:
import torch

def relu(x):
    return torch.max(torch.tensor(0.0), x)

def relu_gradient(x):
    return torch.where(x > 0, torch.tensor(1.0), torch.tensor(0.0))

points = [0.0, 1.0] 
for point in points:
    x = torch.tensor(point, requires_grad=True, dtype=torch.float)  # Specify dtype=torch.float
    y = relu(x)
    y.backward()
    pytorch_gradient = x.grad.item()
    
    explicit_gradient = relu_gradient(x).item()
    
    print(f"Point: {point}")
    print("PyTorch Gradient:", pytorch_gradient)
    print("Explicit Formula Gradient:", explicit_gradient)
    print()
```

```python
# exercise 4:
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))

def plot_sigmoid_and_gradient():
    x = np.linspace(-10, 10, 100)
    
    sigmoid_values = sigmoid(x)
    
    gradient_values = sigmoid_gradient(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, sigmoid_values, label='Sigmoid Function', color='blue')
    
    plt.plot(x, gradient_values, label='Sigmoid Gradient', linestyle='--', color='red')
    
    plt.title('Sigmoid Activation Function and its Gradient')
    plt.xlabel('Input')
    plt.ylabel('Output')
    
    plt.legend()
    
    plt.grid(True)
    plt.show()

plot_sigmoid_and_gradient()
```

```python
# exercise 5 
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_gradient(x):
    return np.where(x > 0, 1, 0)

def plot_relu_and_gradient():
    x = np.linspace(-10, 10, 100)
    relu_values = relu(x)
    
    gradient_values = relu_gradient(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, relu_values, label='ReLU Function', color='blue')
    
    plt.plot(x, gradient_values, label='ReLU Gradient', linestyle='--', color='red')
    plt.title('ReLU Activation Function and its Gradient')
    plt.xlabel('Input')
    plt.ylabel('Output')
    
    plt.legend()
    
    plt.grid(True)
    plt.show()

plot_relu_and_gradient()
```

```python
# exercise 6:
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_gradient(x):
    return 1 - np.tanh(x)**2

def plot_tanh_and_gradient():
    x = np.linspace(-10, 10, 100)
    
    tanh_values = tanh(x)
    
    gradient_values = tanh_gradient(x)
    
    plt.figure(figsize=(10, 6))
    
    plt.plot(x, tanh_values, label='tanh Function', color='blue')
    
    plt.plot(x, gradient_values, label='tanh Gradient', linestyle='--', color='red')
    
    plt.title('Hyperbolic Tangent (tanh) Activation Function and its Gradient')
    plt.xlabel('Input')
    plt.ylabel('Output')
    
    plt.legend()
    
    plt.grid(True)
    plt.show()

plot_tanh_and_gradient()
```

```python
# exercise 7:
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_gradient(x, alpha=0.1):
    return np.where(x > 0, 1, alpha)

def plot_leaky_relu_and_gradient():
    x = np.linspace(-10, 10, 100)
    
    leaky_relu_values = leaky_relu(x)
    
    gradient_values = leaky_relu_gradient(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, leaky_relu_values, label='Leaky ReLU Function', color='blue')
    
    plt.plot(x, gradient_values, label='Leaky ReLU Gradient', linestyle='--', color='red')
    
    plt.title('Leaky ReLU Activation Function and its Gradient')
    plt.xlabel('Input')
    plt.ylabel('Output')
    
    plt.legend()
    
    plt.grid(True)
    plt.show()

plot_leaky_relu_and_gradient()
```

```python
# exercise 8:
import torch
import torch.nn as nn
import torch.optim as optim

X_train = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]]).float()
y_train = torch.tensor([[0], [1], [1], [1]]).float()

class Perceptron(nn.Module):
    def __init__(self, input_dim):
        super(Perceptron, self).__init__()
        self.linear = nn.Linear(input_dim, 1)

    def forward(self, x):
        x = self.linear(x)
        return x

input_size = 2
perceptron = Perceptron(input_size)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(perceptron.parameters(), lr=0.1)

epochs = 1000
for epoch in range(epochs):
    outputs = perceptron(X_train)
    loss = criterion(outputs, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

with torch.no_grad():
    predicted = (torch.sigmoid(perceptron(X_train)) > 0.5).float()
    accuracy = (predicted == y_train).sum().item() / y_train.size(0)
    print(f'Accuracy: {accuracy * 100:.2f}%')
```
